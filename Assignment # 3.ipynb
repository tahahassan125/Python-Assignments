## Q1. Make a calculator using Python with addition , subtraction , multiplication , division and power.

```python
print("\tSelect operation.\n")
print("\ta.for Add")
print("\ts.for Subtract")
print("\tm.for Multiply")
print("\td for.Divide")
print("\tp for.Power\n")
choice = input("Enter choice(a/s/m/d/p): ")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
power = num1 ** num2

if choice == 'a':
   print(num1,"+",num2,"=", addition)

elif choice == 's':
   print(num1,"-",num2,"=", subtraction)

elif choice == 'm':
   print(num1,"*",num2,"=", multiplication)

elif choice == 'd':
   print(num1,"/",num2,"=", division)

elif choice == 'p':
   print(num1,"^",num2,"=", power)

else:
   print("Invalid input")
```

## Q5. Write a program to identify duplicate values from list

```python
list1 = [2, 3, 4, 5, 4, 2]
unique = []
for ele in list1:
    if ele not in unique:
        unique.append(ele)
print(unique)
```
